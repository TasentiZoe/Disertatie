```r
log_likelihood <- logLik(model1)
log_likelihood
```

```r
num_params <- length(coef(model1))
num_obs <- nrow(train_data) 
```

```r
AIC <- -2 * log_likelihood + 2 * num_params
AIC
```

```r
BIC <- -2 * log_likelihood + num_params * log(num_obs)
BIC
```

```r
# Extragei valorile AIC i BIC
aic <- model1$selection[1, "AIC"]
bic <- model1$selection[1, "BIC"]

# Afiai valorile AIC i BIC
cat("AIC:", aic, "\n")
cat("BIC:", bic, "\n")
```
